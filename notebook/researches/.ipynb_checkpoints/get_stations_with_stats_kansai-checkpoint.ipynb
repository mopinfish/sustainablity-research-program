```python
import networkx as nx
import osmnx as ox
import requests
import sys,os,os.path
import pandas as pd
import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ox.config(use_cache=True, log_console=True)
ox.__version__
```

```python
stations = pd.read_csv('../../data/stations.csv')
```

```python
stations.head()
```

```python
target_pref_cds = [
    25, # 滋賀
    26, # 京都
    27, # 大阪
    28, # 兵庫
    29, # 奈良
    30, # 和歌山    
]
target_stations = stations[stations['pref_cd'].isin(target_pref_cds)]

len(target_stations)
```

```python
# 欠損値になりうる駅を削除
len(target_stations)
target_stations = target_stations[target_stations['station_name'] != '保津峡']
len(target_stations)
```

```python
#駅の800m範囲内のグラフ構造の配列を取得
def get_stats(data):
    wurster_hall = (data['lat'], data['lon'])
    one_mile = 800 #meters
    try:
        G = ox.graph_from_point(wurster_hall, dist=one_mile, network_type='drive')

        # what sized area does our network cover in square meters?
        G_proj = ox.project_graph(G)
        nodes_proj = ox.graph_to_gdfs(G_proj, edges=False)
        graph_area_m = nodes_proj.unary_union.convex_hull.area

        # show some basic stats about the network
        stats = ox.basic_stats(G_proj, area=graph_area_m, clean_intersects=True, circuity_dist='euclidean')
        return pd.Series(stats)
    except:
        print(data['station_name'])
```

```python
#駅の800m範囲内のグラフ構造の配列を取得
stats_list = target_stations.apply(get_stats, axis=1)
stats_list
```

```python
target_stations_with_stats = pd.concat([target_stations, stats_list], axis=1)
```

```python
target_stations_with_stats.to_csv('./data/target_kansai_stations_with_stats.csv')
```
