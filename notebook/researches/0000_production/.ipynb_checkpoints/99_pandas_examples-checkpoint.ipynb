```python
%%time

## cluster coefficient from osmnx graph

# density ------------------------------------------------------------------------------------
def calc_density(G):
    key = 'density'
    density_val = round(nx.density(G),8)

    return density_val

# cluster coefficient-------------------------------------------------------------------------
def calc_cluster_coefficient(G):
    key = 'cluster_coefficient'
    G2 = nx.Graph(G)
    cluster_coefficient = nx.clustering(G2, weight='length')

    return cluster_coefficient

# MultiDiGraphでは計算できないのでDiGraphに変換
# MultiDiGraph edge num: 10515
# DiGraph      edge num: 10441 
# average cluster coefficient-----------------------------------------------------------------
def calc_avg_cluster_coefficient(G):
    key = 'average_cluster_coefficient'

    G2 = nx.DiGraph(G)

    # G2_edges = set(G2.edges())
    # for edge in set(G.edges()):
    #     if edge not in G2_edges:
    #         print(edge)

    avg_cluster_coefficient = round(nx.average_clustering(G2, weight='length'), 7)

    return avg_cluster_coefficient

wurster_hall = (35.68195080717844, 139.76703896931784)
one_mile = 800 #meters
G = ox.graph_from_point(wurster_hall, dist=one_mile, network_type='drive')

print(calc_density(G))
print(calc_cluster_coefficient(G))
print(calc_avg_cluster_coefficient(G))
        
```
